SPLITTING OF DATA SET DEPENDENT AND INDEPENDENT VARIABLE 

```python
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values
print(X)
print(Y)
```

FEATURE SCALING OF INDEPENDENT VARIABLE

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
scaler = preprocessing.StandardScaler()
scaled_df = scaler.fit_transform(df.iloc[:, :-1].values)
print(scaled_df)
```

SPLITTING OF DATA TO TRAIN N TEST DATA

```python
target=df['Default']
target.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_df,target,test_size=0.3,random_state=0)
```

TRAIN DATA SET USING LOGISTIC REGRESSION AND  PREDICT TEST DATA SET

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(solver='lbfgs').fit(x_train,y_train)
predictions=model.predict(x_test)
print(predictions)
```

CONFUSION MATRIX GENERATION

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix

confusion_matrix = confusion_matrix(y_test,predictions)
print(confusion_matrix)
```

METRICS CALCULATION

```python
from sklearn.metrics import classification_report
print("Accuracy:",metrics.accuracy_score(y_test,predictions))
print("Precision:",metrics.precision_score(y_test,predictions))
print("Recall:",metrics.recall_score(y_test,predictions))
print(classification_report(y_test,predictions))
```
